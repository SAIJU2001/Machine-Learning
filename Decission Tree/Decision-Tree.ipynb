--- Implement Decision Tree Classification Algorithm ---

```python
#choose independent and dependent variable
#here we are predicting a pokemon is legendary or not, using its speed
x = poke[['Speed']]
y = poke[['Legendary']]
```

```python
pip install scikit-learn
```

```python
#library import for split the data
from sklearn.model_selection import train_test_split
```

```python
#split the data for training and testing
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.3)
```

```python
#we can do both regression and classification using the decission tree
#here use classification
from sklearn.tree import DecisionTreeClassifier
```

```python
#crete the instance
dtc = DecisionTreeClassifier()
```

```python
#fit the model to the instance
dtc.fit(x_train, y_train)
```

```python
#now predict the values against the test data set
y_pred = dtc.predict( x_test)
```

```python
#import metrics to calculate the error of model
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix( y_test, y_pred)
```
